```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

image = mpimg.imread('test6.jpg')
plt.imshow(image)
plt.show()
```

# Convert RGB 2 GRAY

```python
# 1. conver image to gray scale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
binary = np.zeros_like(gray)
binary[(gray > thresh[0]) & (gray <= thresh[1])] = 1

f, (ax1, ax2) = plt.subplots(1,2, figsize=(20,10))
ax1.imshow(gray, cmap='gray')
ax1.set_title('Gray', fontsize=30)
ax2.imshow(binary, cmap='gray')
ax2.set_title('Gray binary', fontsize=30)
plt.show()
```

```python
R = image[:,:,0]
G = image[:,:,1]
B = image[:,:,2]

f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(20,3))
ax1.imshow(R, cmap='gray')
ax1.set_title('R', fontsize=20)

ax2.imshow(G, cmap='gray')
ax2.set_title('G', fontsize=20)

ax3.imshow(B, cmap='gray')
ax3.set_title('B', fontsize=20)

plt.show()
```

```python
thresh = (200, 255)
def binary(img, thresh, title):
    thresh = thresh
    binary = np.zeros_like(img)
    binary[(img > thresh[0]) & (img <= thresh[1])] = 1

    f, (ax1, ax2) = plt.subplots(1,2, figsize=(20,10))
    ax1.imshow(img, cmap='gray')
    ax1.set_title(title, fontsize=30)
    ax2.imshow(binary, cmap='gray')
    ax2.set_title(title+' binary', fontsize=30)
    plt.show()
```

```python
binary(R, thresh, 'R')
```

```python
binary(G, thresh, 'G')
```

```python
binary(B, thresh, 'B')
```

## 2. Convert RGB 2 HLS/Threshold

```python
hls = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
H = hls[:,:,0]
L = hls[:,:,1]
S = hls[:,:,2]

R = image[:,:,0]
G = image[:,:,1]
B = image[:,:,2]

f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(20,3))
ax1.imshow(H, cmap='gray')
ax1.set_title('H', fontsize=20)

ax2.imshow(L, cmap='gray')
ax2.set_title('L', fontsize=20)

ax3.imshow(S, cmap='gray')
ax3.set_title('S', fontsize=20)

plt.show()
```

```python
thresh = [100, 255]
binary(H, thresh, 'H')
```

```python
binary(L, thresh, 'L')
```

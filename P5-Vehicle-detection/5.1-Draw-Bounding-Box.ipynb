```python
# Draw bounding boxes 
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
% matplotlib inline

path = 'test_images/bbox-example-image.jpg'
image = mpimg.imread(path)

def draw_boxes(img, bboxes, color=(0,255,0), thick=6):
    # make a copy of the image
    draw_img = np.copy(img)
    # draw bounding boxes
    for bbox in range(len(bboxes)):
        (x1, y1), (x2, y2) = bboxes[bbox]
        cv2.rectangle( draw_img, (x1, y1), (x2, y2), color, thick)
        # (x1, y1) lower right corner
        # (x2, y2) upper left corner
    return draw_img
```

```python
bboxes = [((830, 500), (1150, 700)), ((400, 580), (270, 480))]
```

```python
result = draw_boxes(image, bboxes)
plt.imshow(result)
plt.show()
```
